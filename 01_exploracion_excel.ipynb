```python
import pandas as pd

ruta = "Austin-Johns-SHINY-Pokemon-Home-Organizer-1.4-PLZA - Dex.csv"

df_dex = pd.read_csv(ruta)

df_dex.head(10)
```

```python
df_dex.columns
```

```python
# 1. Creamos una copia limpia para no alterar el dataframe original
df_catalogo = df_dex.copy()

# 2. Creamos nuestro 'id_home' usando el índice de las filas (+1 para que empiece en 1 y no en 0)
df_catalogo['id_home'] = df_catalogo.index + 1

# 3. Selecciona y renombra solo las columnas que te interesan del Excel de Austin John.
# (Ajusta los nombres 'Nombre', 'Número' etc., según cómo se llamen exactamente en tu dataframe)
# Aquí un ejemplo asumiendo nombres estándar:
df_catalogo = df_catalogo[['id_home', 'Nat', 'Keyword', '-Form']]

# 4. Renombramos a minúsculas y sin espacios (Estilo profesional para bases de datos)
df_catalogo.columns = ['id_home', 'numero_pokedex', 'nombre', 'forma_variante']

# Vemos cómo ha quedado estructurado tu catálogo
df_catalogo.head()
```

```python
ddd = df_catalogo.set_index('id_home')

ddd.tail(20)
```

```python
df_catalogo_limpio = df_catalogo.iloc[:1387]

df_catalogo_limpio.tail(3)
```

```python
import sqlite3

# 1. Creamos la conexión. Si el archivo 'shinydex.db' no existe, SQLite lo creará automáticamente.
conexion = sqlite3.connect("shinydex.db")

# 2. Volcamos el DataFrame a una tabla de SQL llamada 'catalogo_dex'
df_catalogo_limpio.to_sql(
    name='catalogo_dex', 
    con=conexion, 
    if_exists='replace', # Si volvemos a ejecutar la celda, reemplaza la tabla para no duplicar
    index=False          # No guardamos el índice automático de Pandas, ya creamos 'id_home'
)

# 3. Cerramos la conexión por seguridad
conexion.close()

print("¡Base de datos creada y catálogo importado con éxito!")
```
