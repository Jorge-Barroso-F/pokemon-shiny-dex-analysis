```python
import pandas as pd

ruta_dex = "Austin-Johns-SHINY-Pokemon-Home-Organizer-1.4-PLZA - Dex.csv"
ruta_imp_idx = "Austin-Johns-SHINY-Pokemon-Home-Organizer-1.4-PLZA - Imported Index.csv"

df_dex = pd.read_csv(ruta_dex)
df_imp_idx = pd.read_csv(ruta_imp_idx)

df_imp_idx_2 = df_imp_idx[~df_imp_idx['S-LivingForm'].str.contains('gen', case=False, na=False)].reset_index(drop=True)

df_dex_c = df_dex.copy()
df_imp_idx_c = df_imp_idx_2.copy()[['S-LivingForm', 'Unnamed: 1']]

df_dex_c['id_home'] = df_dex_c.index + 1
df_imp_idx_c['id_home'] = df_imp_idx_c.index + 1
```

```python
df_dex_c = df_dex_c.set_index('id_home')
df_imp_idx_c = df_imp_idx_c.set_index('id_home')
```
